##### Consider the two dimensional data matrix [(2, 1), (3, 4), (5, 0), (7, 6), (9, 2)].Implement principal component analysis. Use this to obtain the feature in transformed 2D feature space. Plot the scatter plot of data points in both the original as well as transformed domain.

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = np.array([(2, 1), (3, 4), (5, 0), (7, 6), (9, 2)])
```

```python
mean = np.mean(data, axis=0)
cov_matrix = np.cov(data.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
sorted_indices = np.argsort(eigenvalues)[::-1]
principal_components = eigenvectors[:, [sorted_indices[0], sorted_indices[1]]]
```

```python
transformed_data = np.dot(data - mean, principal_components)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], color='green')
plt.title('Original Data')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.subplot(1, 2, 2)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], color='purple')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
```

##### Implement Principal Component Analysis Algorithm and use it to reduce dimensions of Iris Dataset (from 4D to 2D). Plot the scatter plot for samples in the transformed domain with different colour codes for samples belonging to different classes.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
mean = np.mean(data, axis=0)
cov_matrix = np.cov(data.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
sorted_indices = np.argsort(eigenvalues)[::-1]  
principal_components = eigenvectors[:, sorted_indices[:2]]
```

```python
transformed_data = np.dot(data - mean, principal_components)
```

```python
plt.figure(figsize=(8, 6))
colors = ['pink', 'yellow', 'orange']
for i in range(3):
    plt.scatter(transformed_data[target == i, 0], transformed_data[target == i, 1], c=colors[i], label=f'Class {i}')
plt.title('Scatter plot of Iris dataset in transformed domain')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
