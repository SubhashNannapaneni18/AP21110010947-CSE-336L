### Import the Iris dataset. Prepare a dataset considering samples belong to any two output classes. Draw the scatter plot for all the samples in the new dataset considering any two input attributes. Examine the scatter plot to find the equation of a line that can separate sample of two classes.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = iris.data
```

```python
target = iris.target
```

```python
df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target
```

```python
class_0_indices = df[df['target'] == 0].index
class_1_indices = df[df['target'] == 1].index
```

```python
new_dataset = pd.concat([df.loc[class_0_indices], df.loc[class_1_indices]])
```

```python
x_attribute = 'sepal length (cm)'
y_attribute = 'sepal width (cm)'
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(new_dataset[new_dataset['target'] == 0][x_attribute],
            new_dataset[new_dataset['target'] == 0][y_attribute],
            label='Class 0', color='purple')
plt.scatter(new_dataset[new_dataset['target'] == 1][x_attribute],
            new_dataset[new_dataset['target'] == 1][y_attribute],
            label='Class 1', color='pink')
plt.xlabel(x_attribute)
plt.ylabel(y_attribute)
plt.title('Scatter Plot of Iris Dataset (Classes 0 and 1)')
plt.legend()
plt.grid(True)
```

```python
plt.plot([4.5, 7.5], [2, 4.5], linestyle='--', color='green', label='Decision Boundary')

plt.legend()
plt.show()
```

### Take two 3D coordinates from the user. Find out the distance between these points using following distance functions. x and y are vectors representing the coordinates of two points.Here, k=3

```python
import numpy as np
```

```python
import math
```

```python
def euclidean_distance(point1, point2):
    return math.sqrt(sum([(x - y) ** 2 for x, y in zip(point1, point2)]))
```

```python
def manhattan_distance(point1, point2):
    return sum([abs(x - y) for x, y in zip(point1, point2)])
```

```python
def minkowski_distance(point1, point2, p=2):
    return sum([(abs(x - y)) ** p for x, y in zip(point1, point2)]) ** (1 / p)
```

```python
point1 = []
point2 = []
```

```python
print("Enter the coordinates for the first point:")
for i in range(3):
    point1.append(float(input(f"Enter coordinate {i + 1}: ")))
```

```python
print("\nEnter the coordinates for the second point:")
for i in range(3):
    point2.append(float(input(f"Enter coordinate {i + 1}: ")))
```

```python
euclidean_dist = euclidean_distance(point1, point2)
manhattan_dist = manhattan_distance(point1, point2)
minkowski_dist = minkowski_distance(point1, point2, p=3)
```

```python
print("\nEuclidean Distance:", euclidean_dist)
print("Manhattan Distance:", manhattan_dist)
print("Minkowski Distance (p=3):", minkowski_dist)
```
